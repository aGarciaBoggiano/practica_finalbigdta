```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra

# B. Una función que calcule el precio total de la compra con un descuento del 10%

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

# D. Una función que calcule el precio total aplicando el IVA (21%)

# E.lista los artículos cuyo precio es superior a 20€
```

```python
# A. Una función que calcule el precio total de la compra
get_value = lambda x: articulos[x]
total_compra =list(map(get_value,compra))
total = reduce(lambda x, y:x+y,total_compra)
print(total)
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%
get_value = lambda x: articulos[x]
total_compra =list(map(get_value,compra))
total = reduce(lambda x, y:x+y,total_compra)

def calcular_descuento(porcentaje,total ):
    return total * porcentaje / 100

cd=calcular_descuento(10,total)

totaldescuento=(total-cd)
print(totaldescuento)
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
get_value = lambda x: articulos[x]
total_compra =list(map(get_value,compra))
total = reduce(lambda x, y:x+y,total_compra)

def calcular_descuento(porcentaje,total ):
    return total * porcentaje / 100

cd=calcular_descuento(10,total>=100)

totaldescuento=(total-cd)
print(totaldescuento)
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%)
get_value = lambda x: articulos[x]
total_compra =list(map(get_value,compra))
total = reduce(lambda x, y:x+y,total_compra)

def calcular_impuesto(porcentaje,total ):
    return total * porcentaje / 100

cp= calcular_impuesto(21,total)

totalimpuesto=(total+cp)
print(totalimpuesto)
```

```python
# E.lista los artículos cuyo precio es superior a 20€
articulo_20=dict(filter(lambda x:x[1]>20,articulos.items()))
print(articulo_20)
```
